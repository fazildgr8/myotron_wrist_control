## Make windowed Data

```python
def make_windowed(data,win_len):
    nb = data.shape[0]
    X = []
    for i in tqdm(range(nb-win_len)):
        X.append(data[i:i+win_len].values)
    return np.array(X)
```

```python
emg_df = df[emg_labels]
```

```python
emg_df.shape
```

```python
windowed_test = make_windowed(emg_df,250)
```
