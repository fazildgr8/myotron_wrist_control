```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import glob
import warnings
import os
from tqdm import tqdm
# warnings.filterwarnings(action='once')
# warnings.filterwarnings('ignore')
# warnings.simplefilter('ignore')
```

## Import the Datasets to Dataframe

```python
win_len = 250
```

```python
X = np.load('prepared_data/X_all_ProSup_{}_40.npy'.format(win_len))
y = np.load('prepared_data/Y_hot_all_ProSup_{}_40.npy'.format(win_len))
```

```python
X = np.reshape(X,(X.shape[0],X.shape[1],X.shape[2]))
X.shape
```

```python
y.shape
```

## Convert to Binary

```python
new_y = []
for d in y:
    if(d[0]==1 or d[2]==1):
        new_y.append([1,0])
    if(d[1]==1 or d[3]==1):
        new_y.append([0,1])
y = np.array(new_y)
y.shape
```

## Convert to Thirtiary

```python
new_y = []
for d in y:
    if(d[0]==1):
        new_y.append([1,0,0])
    if(d[1]==1 or d[3]==1):
        new_y.append([0,1,0])
    if(d[2]==1 or d[4]==1):
        new_y.append([0,0,1])
y = np.array(new_y)
y.shape
```

## Time Series Data Augumentation

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

### Merge, Shuffle and Split

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True)
```

##### Keep only 8 EMG data

```python
from tqdm.notebook import trange, tqdm
def rm_columns(X,n):
#     X_r = list(np.zeros((X.shape[0],X.shape[1],X.shape[2]-n)))
    X_r = []
    for i in tqdm(range(len(X))):
        d = np.copy(X[i])
        for i in range(n):
            d = np.delete(d, -1, 1)
        X_r.append(d)
    return np.array(X_r)
```

```python
X_train = rm_columns(X_train,4)
X_test = rm_columns(X_test,4)
```

## Distribution of Labels

```python
def countX(lst, x): 
    count = 0
    for ele in lst: 
        if (ele == x): 
            count = count + 1
    print(count)
    return count
```

### Test Distribution

```python
sum = 0
for i in range(3):
    c = countX(y_true,i)
    sum = sum+c/len(y_true)*100
    print(i,'=',c/len(y_true)*100,' %')
print('All ',sum,' %')
```

### Train Distribution

```python
sum = 0
for i in range(3):
    c = countX(y_true_train,i)
    sum = sum+c/len(y_true_train)*100
    print(i,'=',c/len(y_true_train)*100,' %')
print('All ',sum,' %')
```

## Rest Class Bias Correction

```python
cnt_train = 140000
cnt_test = 40000
new_X_train = []
new_y_train = []
new_X_test = []
new_y_test = []
for i in tqdm(range(len(X_train))):
    if(cnt_train>=0 and y_train[i][0]==1):
        new_X_train.append(X_train[i])
        new_y_train.append(y_train[i])
        cnt_train=cnt_train-1
    if(y_train[i][1]==1 or y_train[i][2]==1):
        new_X_train.append(X_train[i])
        new_y_train.append(y_train[i])
X_train = np.array(new_X_train)
y_train = np.array(new_y_train)

for i in tqdm(range(len(X_test))):
    if(cnt_test>=0 and y_test[i][0]==1):
        new_X_test.append(X_test[i])
        new_y_test.append(y_test[i])
        cnt_test=cnt_test-1
    if(y_test[i][1]==1 or y_test[i][2]==1):
        new_X_test.append(X_test[i])
        new_y_test.append(y_test[i])
X_test = np.array(new_X_test)
y_test = np.array(new_y_test)
```

# InceptionTime Trainer

```python
import numpy as np
import sklearn
import pandas as pd
import sys
import subprocess
import tensorflow as tf
from InceptionTime.classifiers.inception import Classifier_INCEPTION
```

### Build New Model

```python
 clf = Classifier_INCEPTION('', (X_train.shape[1],X_train.shape[2]),nb_classes=3,
                            verbose=True,batch_size=250,nb_epochs=25,nb_filters=16,
                            depth=22, kernel_size=11)
```

## Continue Previous Training Session

```python
from keras.models import load_model
clf.model = load_model('best_model.hdf5')
clf.model.summary()
```

```python
df_metrics = clf.fit(X_train, y_train, X_test, y_test, y_true,plot_test_acc=True)
```
